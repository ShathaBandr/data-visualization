```python
import matplotlib.pyplot as plt 
years = [2020, 2021, 2022, 2023, 2024, 2025]
sales = [200, 400, 600, 800, 1200, 1400]

plt.plot(years, sales, marker='o', linestyle='-', color='b', label="Sales")

plt.xlabel('Year')
plt.ylabel('Sales in (KSA)')
plt.title('yearly sales Growth')
plt.legend()

plt.show()
```

```python
categories = ['Electronics', 'Clothing', 'Home Decor', 'Books']
sales = [500, 600, 700, 800]

plt.bar(categories, sales, color=['blue', 'pink', 'green', 'brown'])

plt.xlabel('Product Category')
plt.ylabel('Sales (in KSA)')
plt.title('Sales by category')

plt.show()
```

```python
import seaborn as sns 
import pandas as pd 

data = {
    'category': ['Electronics', 'Clothing', 'Home Decor', 'Books'], 
    'sales': [500, 600, 700, 800]
}

df = pd.DataFrame(data)

sns.barplot(x='category', y='sales', data=df, palette='coolwarm')

plt.title("Sales by category (Seaborn)")

plt.show()
```

```python
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, color='green', edgecolor='grey')

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("histogram of random data")

plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color="pink")

plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Scatter plot example')

plt.show()
```
